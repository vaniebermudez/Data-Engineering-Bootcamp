```python
# Excel Sheet -> Python Reads or Ingests -> Pandas organizes it

import pandas as pd
import numpy as np
```

```python
# DataFrame
data = {
    "product": ['Laptop', 'Mouse', 'Keyboard', 'Monitor'],
    "price": [1200.50, 25.00, 45.99, 3500.00],
    "quantity": [2, 20, 10, 5]
}

df = pd.DataFrame(data)
df
```

```python
# Reading data from CSV
df_sales = pd.read_csv('raw_sales_data.csv')
df_sales
```

```python
df_sales.head()
```

```python
df_sales['Amount']
```

```python
# Filtering
df_sales[(df_sales["Amount"]> 1000) & (df_sales["customer_id"]=="C001")]
```

```python
# Calculations
df['total_sales'] = df["price"] * df['quantity']
df
```

```python
# Group By
ts = df.groupby('product')['total_sales'].sum()
```

```python
# Sort Values
ts = ts.sort_values(ascending=False)
ts
```

```python
# Output as CSV file
ts.to_csv('sales_by_product.csv')
```
