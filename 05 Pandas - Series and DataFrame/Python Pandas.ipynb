Pandas - Series

```python
sales_index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']
sales_values = [500, 200, 300, 200, 100]

sales_series = pd.Series(sales_values, index=sales_index)
sales_series
```

```python
sales = {
    'Monday': 500,
    'Tuesday': 200,
    'Wednesday': 300,
    'Thursday': 150,
    'Friday': 175
}

sales_series_dict = pd.Series(sales)
sales_series_dict
```

```python
# Acccessing Data

# []
sales_series_dict[1]
sales_series_dict['Tuesday']
```

```python
# [[]]
sales_series_dict[[0]]
```

```python
# [:]
sales_series_dict[:3]
```

```python
# [[n, n, n]]

sales_series_dict[[0, 2, 4]]
```

```python
# [condition]
sales_series_dict[sales_series_dict>=300]
```

Pandas - DataFrame

```python
sales_month = {
    'Monday': [200, 250, 120, 500],
    'Tuesday': [100, 200, 300, 400],
    'Wednesday': [100, 150, 160, 109],
    'Thursday': [530,520,106,806],
    'Friday': [900, 899, 121, 949]
}

week_index = ['week1', 'week2', 'week3', 'week4']
test_series = pd.DataFrame(sales_month, index=week_index)
test_series['weekly_total'] = test_series.sum(axis=1)

test_series
```

```python
test_series.T
```

```python
# iloc

test_series.iloc[2:3, 3:4]
```

```python
# loc
test_series.loc[['week1', 'week4'], ['Monday', 'Wednesday']]
```

```python
new_series = test_series.T
new_series
```

```python
new_series.loc['Saturday'] = [100, 200, 300, 400]
new_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']

new_series = new_series.reindex(new_order)
new_series
```

```python
new_series.loc['weekly_total'] = new_series.sum(axis=0)
new_series
```

```python
new_series.drop('weekly_total')
new_series
```

```python
clean_sales = sales.fillna(sales.mean())
clean_sales
```
